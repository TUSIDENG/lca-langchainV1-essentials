What is Jupyter?

Jupyter is an open-source interactive computing environment for writing and running Jupyter Notebook documents (with the .ipynb extension), which contain live code, textual explanations, mathematical formulas, charts, and more. It is widely used in data analysis, machine learning, scientific computing, and teaching. Its core advantage is its **"code-as-you-write"** interpretation feature, which allows you to run code and view results in real time, making it ideal for exploratory programming and presentation.

Core features of Jupyter:

1. Multi-language support: Although Python is the most commonly used, multiple languages (such as R, Julia, SQL, etc.) are supported through the "kernel", which is also the origin of its name "Jupyter" (Julia + Python + R).

2. Interactive Programming: Code can be run in sections (by cell), and the results (such as output and charts) are displayed directly below the code, making it easier to debug and verify.

3. Rich Text Display: Supports Markdown-formatted text, LaTeX mathematical formulas (e.g., E=mc2), images, HTML, and more, seamlessly integrating code and documentation.

4. Cross-platform: It can run on Windows, macOS, and Linux, and also supports remote access to the Jupyter service on the server through a browser.
